```python
pd.set_option('display.max_columns', None)
```

```python
cols = ['id', 'diagnosis', 'radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1', 'compactness1', 'concavity1', 'concave_points1', 'symmetry1', 'fractal_dimension1', 'radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2', 'compactness2', 'concavity2', 'concave_points2', 'symmetry2', 'fractal_dimension2', 'radius3', 'texture3', 'perimeter3', 'area3', 'smoothness3', 'compactness3', 'concavity3', 'concave_points3', 'symmetry3', 'fractal_dimension3',]
```

```python
len(cols)
```

```python
df = pd.read_csv('wdbc.data', names=cols)
df.head()
```

```python
y = df.iloc[:, 1].values
```

```python
y 
```

```python
df.drop(['id'], axis=1)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
df = df.drop(['diagnosis', 'id'], axis=1)
```

```python
df.head()
```

```python
X = df.iloc[:, :].values
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
y_test = y_test.reshape(len(y_test), 1)
```

```python
y_test
```

```python
classifier = LogisticRegression(random_state=0)
```

```python
classifier.fit(X_train, y_train);
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
np.set_printoptions(precision=3)
np.concatenate((y_test, y_pred.reshape(len(y_pred), 1)),1)
```

```python
matrix = confusion_matrix(y_test, y_pred)
```

```python
matrix
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
score * 100 
```

```python
random_forest = RandomForestClassifier(random_state=0, n_estimators=100, criterion='entropy')
```

```python
X_train
```

```python
y_train
```

```python
random_forest.fit(X_train, y_train)
```

```python
forest_pred = random_forest.predict(X_test)
forest_pred
```

```python
np.set_printoptions(precision=3)
np.concatenate((y_test, forest_pred.reshape(len(forest_pred), 1)), 1)
```

```python
confusion_matrix(y_test, forest_pred)
```

```python
accuracy_score(y_test, forest_pred) * 100
```

## Logistic Regression gives us 97 % accuracy whereas Random Forest gave us 95 % accuracy. Thus, Logistic regression is better for this particular dataset.
